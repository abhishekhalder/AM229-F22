# Problem 1. [50 points] Affine and Convex Sets, Cones, and Hulls

## (a) [10 + (1+4) = 15 points] Copositive matrices 

(i) In Lec. 4, p. 3, we mentioned without proof that the set of copositive matrices
$$\mathbb{S}_{\rm{copo}}^{n} := \{X\in\mathbb{S}^{n} \mid x^{\top}X x \geq 0 \;\text{for all}\;x\in\mathbb{R}^{n}_{+}\}$$
is convex. Please note there was a typo (now corrected in notes) in the bottom of that page: the correct set inclusion is: $\mathbb{S}^{n}_{+} \subset \mathbb{S}^{n}_{\rm{copo}} \subset \mathbb{S}^{n}$.

**Prove that** $\mathbb{S}_{\rm{copo}}^{n}$ is a **convex set**.

(ii) Is $\mathbb{S}_{\rm{copo}}^{n}$ also a **cone**? **Why/why not?**

## Solution for Prob. 1(a).

(i) For all $X_1,X_2\in \mathbb{S}_{\rm{copo}}^{n}$, we have $x^{\top}X_i x \geq 0$ for all $x\in\mathbb{R}^{n}_{+}$, $i\in\{1,2\}$. 

Therefore, for all $0\leq \theta\leq 1$, we have $x^{\top}\left(\theta X_1 + (1-\theta)X_2\right)x = \theta x^{\top}X_1 x + (1-\theta)x^{\top}X_1 x \geq 0$ for all $x\in\mathbb{R}^{n}_{+}$, i.e., $\theta X_1 + (1-\theta)X_2\in \mathbb{S}_{\rm{copo}}^{n}$. Hence, $\mathbb{S}_{\rm{copo}}^{n}$ is a convex set.

(ii) Yes, $\mathbb{S}_{\rm{copo}}^{n}$ is a cone because if $X\in \mathbb{S}_{\rm{copo}}^{n}$, then $x^{\top}\theta X x \geq 0$ for all $x\in\mathbb{R}^{n}_{+}$ for all $\theta \geq 0$, i.e., $\theta X\in \mathbb{S}_{\rm{copo}}^{n}$ for all $\theta \geq 0$.

## (b) [(1+9) + (1+4) = 15 points] Permutation matrices

(i) A permutation matrix of size $n\times n$ is a binary matrix such that it has exactly single 1 in each row and each column, the rest of the entries are zero. For example, there are only two permuation matrices of size $2\times 2$: 
$$\begin{pmatrix}1 & 0\\
0 & 1\end{pmatrix}, \quad \begin{pmatrix}0 & 1\\
1 & 0\end{pmatrix}.$$
In general, there are $n!$ permutation matrices of size $n\times n$. For fixed $n$, is the set of all $n\times n$ permutation matrices a **convex set**? **Why/why not?**

(ii) For fixed $n$, is the set of all $n\times n$ permutation matrices a **cone**? **Why/why not?**

## Solution for Prob. 1(b).

(i) No, the set of all $n\times n$ permutation matrices for fixed $n$, is NOT convex. As a counterexample, consider $n=2$, and the convex combination 
$$\theta\begin{pmatrix}1 & 0\\
0 & 1\end{pmatrix} + (1-\theta)\begin{pmatrix}0 & 1\\
1 & 0\end{pmatrix} = \begin{pmatrix}\theta & 1-\theta\\
1-\theta & \theta\end{pmatrix},$$
which is NOT a permutation matrix for all $0\leq \theta\leq 1$. So the set of permutation matrices is nonconvex.

(ii) No, the set of all $n\times n$ permutation matrices for fixed $n$, is NOT a cone. As a counterexample, consider $n=2$, and the permutation matrix $I_2$ (the $2\times 2$ identity matrix). The matrix $\theta I_{2}$ is NOT a permutation matrix for all $\theta\geq 0$.


## (c) [(1+4) + (1+9) + (1+4) = 20 points] Probability simplex

(i) Consider the probability simplex $\Delta^{n-1}\subset\mathbb{R}^{n}$ introduced in Lec. 2, p. 5-6. What is the **affine hull** ${\rm{aff}}(\Delta^{n-1})$? **Explain briefly**.

(ii) Is $\Delta^{n-1}$ a **convex set**? **Why/why not?**

(iii) Is $\Delta^{n-1}$ a **cone**? **Why/why not?**

## Solution for Prob. 1(c).

(i) The affine hull
$${\rm{aff}}(\Delta^{n-1}) = {\rm{aff}}\{\boldsymbol{x}\in\mathbb{R}^{n}_{+} \mid \boldsymbol{1}^{\top}\boldsymbol{x} = 1\} = \bigg\{\sum_{i=1}^{k}\theta_i \boldsymbol{x}_i \mid \boldsymbol{x}_i\in\Delta^{n-1},\sum_{i=1}^{k}\theta_i = 1,\theta_i\in\mathbb{R}\,\text{for all}\,i=1,...,k\bigg\} = \{\boldsymbol{x}\in\mathbb{R}^{n}\mid \boldsymbol{1}^{\top}\boldsymbol{x} = 1\},$$
i.e., the entire hyperplane $\boldsymbol{1}^{\top}\boldsymbol{x} = 1$ imbedded in $\mathbb{R}^{n}$. As expected, $\Delta^{n-1} \subset {\rm{aff}}(\Delta^{n-1}) = \{\boldsymbol{x}\in\mathbb{R}^{n}\mid \boldsymbol{1}^{\top}\boldsymbol{x} = 1\}$.

(ii) Yes, $\Delta^{n-1}$ is a convex set. This is because for any $\boldsymbol{x}_1,\boldsymbol{x}_2\in\Delta^{n-1}$, and any $0\leq \theta \leq 1$, we have $\theta \boldsymbol{x}_1 + (1-\theta)\boldsymbol{x}_2 \in\mathbb{R}^{n}_{+}$. Furthermore,
$$\boldsymbol{1}^{\top}\left(\theta \boldsymbol{x}_1 + (1-\theta)\boldsymbol{x}_2\right) = \theta \left(\underbrace{\boldsymbol{1}^{\top}\boldsymbol{x}_1}_{=1}\right) + (1-\theta)\left(\underbrace{\boldsymbol{1}^{\top}\boldsymbol{x}_2}_{=1}\right) = \theta + (1-\theta) = 1.$$
Therefore, $\theta \boldsymbol{x}_1 + (1-\theta)\boldsymbol{x}_2 \in \Delta^{n-1}$ for any $\boldsymbol{x}_1,\boldsymbol{x}_2\in\Delta^{n-1}$, and any $0\leq \theta \leq 1$. Hence the set $\Delta^{n-1}$ is convex.

(iii) No, $\Delta^{n-1}$ is NOT a cone because for all $\theta\geq 0$ and for all $\boldsymbol{x}\in\Delta^{n-1}$, we have $\theta\boldsymbol{x}\notin\Delta^{n-1}$.

# Problem 2. [15 + 15 + 20 = 50 points] Derivative of Scalar with respect to Matrix

Before attempting the following, please review Lec. 4, p. 12-17. **Compute $\dfrac{\partial f}{\partial X}$ for**

(a) $f(X) = {\rm{tr}}(X^2)$, where $X\in\mathbb{R}^{n\times n}$,

(b) $f(X) = {\rm{tr}}(X^{\top}X)$, where $X\in\mathbb{R}^{m\times n}$,

(c) $f(X) = -\log\det X$, where $X\in\mathbb{S}^{n}_{++}$.

(Hint: use $\det(I + hY) = 1 + h\: {\rm{tr}}(Y) + O(h^2)$, and that $\log(1+y)= y + O(y^2)$ for $|y|$ smaller than $1$)

# Solution for Prob. 2.

(a) The directional derivative
\begin{align*}
D_{Z}f(X) &= \displaystyle\lim_{h\rightarrow 0}\dfrac{{\rm{tr}}((X+hZ)^2) - {\rm{tr}}(X^2)}{h}\\
&= \displaystyle\lim_{h\rightarrow 0}\dfrac{{\rm{tr}}(X^2 + hXZ + hZX + h^2Z^{2}) - {\rm{tr}}(X^2)}{h}\\
&= 2{\rm{tr}}(XZ)\qquad\text{(using invariance of trace under cyclic permutation)}\\
&= {\rm{tr}}\left(\left(2X^{\top}\right)^{\top}Z\right),
\end{align*}
and therefore, $\dfrac{\partial f}{\partial X} = 2X^{\top}$.

(b) The directional derivative
\begin{align*}
D_{Z}f(X) &= \displaystyle\lim_{h\rightarrow 0}\dfrac{{\rm{tr}}((X+hZ)^{\top}(X+hZ)) - {\rm{tr}}(X^{\top}X)}{h}\\
&= \displaystyle\lim_{h\rightarrow 0}\dfrac{{\rm{tr}}(X^{\top}X + hX^{\top}Z + hZ^{\top}X + h^2Z^{\top}Z) - {\rm{tr}}(X^{\top}X)}{h}\\
&= 2{\rm{tr}}(X^{\top}Z)\qquad\text{(using invariance of trace under transposition)}\\
&= {\rm{tr}}\left(\left(2X\right)^{\top}Z\right),
\end{align*}
and therefore, $\dfrac{\partial f}{\partial X} = 2X$.

(c) The directional derivative
\begin{align*}
D_{Z}f(X) &= \displaystyle\lim_{h\rightarrow 0}\dfrac{-\log\det(X+hZ) + \log\det(X)}{h}\\
&= \displaystyle\lim_{h\rightarrow 0}\dfrac{-\log\det(X(I+hX^{-1}Z)) + \log\det(X)}{h}\\
&= \displaystyle\lim_{h\rightarrow 0}\dfrac{-\log\det(X) -\log\det(I+hX^{-1}Z) + \log\det(X)}{h} \quad \text{(using $\det$ of product equals product of $\det$)}\\
&= \displaystyle\lim_{h\rightarrow 0}\dfrac{-\log(1+h{\rm{tr}}(X^{-1}Z))}{h} \quad \text{(using $\det(I + hY) = 1 + h\: {\rm{tr}}(Y) + O(h^2)$)}\\
&= \displaystyle\lim_{h\rightarrow 0}\dfrac{-h{\rm{tr}}(X^{-1}Z)}{h} \quad \text{(using $\log(1+y)= y + O(y^2)$ for $|y|$ smaller than $1$)}\\
&= {\rm{tr}}\left(\left(-X^{-\top}\right)^{\top}Z\right),
\end{align*}
and therefore, $\dfrac{\partial f}{\partial X} = -X^{-\top} = -X^{-1}$ since $X\in\mathbb{S}^{n}_{++}$ (thus symmetric).
