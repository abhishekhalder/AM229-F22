# Problem 1. [10 x 7 = 70 points] Convex Functions

For each of the following, **state clearly if the function is convex, concave, neither convex nor concave**. **Then prove your statement**.

(a) $f(x) = \dfrac{\|Ax + b\|_2^2}{\langle c, x\rangle + d}$ over ${\rm{dom}}(f)=\{x\in\mathbb{R}^{n}\mid \langle c, x\rangle + d > 0\}$ where $A\in\mathbb{R}^{m\times n}$, $b\in\mathbb{R}^{m}$, $c\in\mathbb{R}^{n}$, $d\in\mathbb{R}$. 

(b) $f(X)=\sum_{i=1}^{n}\lambda_{i}(X)\log\lambda_i(X)$ over ${\rm{dom}}(f)=\mathbb{S}^{n}_{++}$ where $\lambda_{i}(X)$ denotes the $i$th eigenvalue of $X$.

(c) $f(x) = \underset{i=1,...,k}{\max}\|A^{(i)}x - b^{(i)}\|_3$ over ${\rm{dom}}(f)=\mathbb{R}^{n}$ where $A^{(i)}\in\mathbb{R}^{m\times n}$, $b^{(i)}\in\mathbb{R}^{m}$ for all $i=1,...,k$.

(d) $f(x) = {\rm{ReLU}}\left(\langle a , x\rangle + b\right)$ over ${\rm{dom}}(f)=\mathbb{R}^{n}$ where $a\in\mathbb{R}^{n}$, $b\in\mathbb{R}$, ${\rm{ReLU}}(y):=\max\{0,y\}$ for $y\in\mathbb{R}$.

(e) $f(x) = -\int_{0}^{2\pi}\log\left(\sum_{k=1}^{n}x_{k}\cos((k-1)\theta)\right){\rm{d}}\theta$ over ${\rm{dom}}(f)=\{x\in\mathbb{R}^{n}\mid \sum_{k=1}^{n}x_{k}\cos((k-1)\theta) > 0 \; \forall \; \theta \in [0, 2\pi)\}$.

(f) $f(X) = \lambda_{\min}(X)$ over ${\rm{dom}}(f)=\mathbb{S}^{n}$ where $\lambda_{\min}(X)$ denotes the minimum eigenvalue of $X$.

(g) $f(x) = \log\left({\rm{logistic}}(x)\right)$ over ${\rm{dom}}(f)=\mathbb{R}$ where ${\rm{logistic}}(x):=\frac{\exp(x)}{1+\exp(x)}$.

## Solution for Prob. 1

(a) First, note that ${\rm{dom}}(f)=\{x\in\mathbb{R}^{n}\mid \langle c, x\rangle + d > 0\}$ is an open halfspace, and hence a **convex set**.

The given function is a compositie map $f = \phi \circ \psi$ where $\phi(y,t) := \frac{y^{\top}y}{t}$ (quadratic-over-linear), and $\psi$ is an affine map given by
$$\begin{pmatrix}
y\\
t
\end{pmatrix} = \psi(x) := \begin{pmatrix}
Ax + b\\
\langle c, x\rangle + d
\end{pmatrix}.$$
From Lec. 8, p. 12, we know that $\phi(y,t)$ is a convex function in $(y,t)\in\mathbb{R}^{n}\times\mathbb{R}_{++}$. From Lec. 9, p. 9, we know that post-composition with affine map preserves function convexity. Therefore, $f(x)$ is a **convex function** over ${\rm{dom}}(f)=\{x\in\mathbb{R}^{n}\mid \langle c, x\rangle + d > 0\}$.

(b) First, note that ${\rm{dom}}(f)=\mathbb{S}^{n}_{++}$ is a **convex set** (in fact, a convex cone) following Lec. 5, p. 12.

Next, let $\phi(x) := x\log x$ for $x\in\mathbb{R}_{++}$. Then $f(X) = \rm{tr}}\left(\phi(X)\right) = \sum_{i=1}^{n}\phi(\lambda_i)$. Notice that $\phi(x)$ is a convex function since $\phi^{\prime\prime}(x) = 1/x > 0$ over its own convex domain ${\rm{dom}}(\phi)=\mathbb{R}_{++}$. Therefore, from $\texttt{Supplementary Notes Operator Convexity and Monotonicity.pdf}$, p. 6, we conclude that $f(X)$ is a **convex function** over ${\rm{dom}}(f)=\mathbb{S}^{n}_{++}$.

(c) First, notice that ${\rm{dom}}(f)=\mathbb{R}^{n}$ is an affine set, and therefore, a **convex set**.

Since $\|\cdot\|_{p}$ is a convex function for any $p\geq 1$ (Lec. 8, p. 11), hence  $\|\cdot\|_{3}$ is a convex function, and its post-composition with affine map preserves function convexity (Lec. 9, p. 9). So $\|A^{(i)}x - b^{(i)}\|_3$ is a convex function for a fixed matrix-vector pair $(A^{(i)},b^{(i)})$. Thus, $f(x)$ is the pointwise max of $k$ convex functions, which by Lec. 9, p. 10, a **convex function** since pointwise max preserves function convexity.

(d) First, notice that ${\rm{dom}}(f)=\mathbb{R}^{n}$ is an affine set, and therefore, a **convex set**.

Since $f(x)$ is pointwise max composed with affine, hence by Lec. 9, p. 9-10, is a **convex function**.

(e) First, notice that ${\rm{dom}}(f)=\{x\in\mathbb{R}^{n}\mid \sum_{k=1}^{n}x_{k}\cos((k-1)\theta) > 0 \; \forall \; \theta \in [0, 2\pi)\}$ is an open halfspace, and hence a **convex set**.

By Lec. 8, p. 10, the function $-\log(x)$ is convex for $x>0$. Hence for a fixed $\theta \in [0, 2\pi)$, the function $-\log\left(\sum_{k=1}^{n}x_{k}\cos((k-1)\theta)\right)$ being convex function post composed with affine, is convex (Lec. 9, p. 9). Therefore, by Lec. 9, p. 8, the nongeative wighted integral of the convex integrand $-\log\left(\sum_{k=1}^{n}x_{k}\cos((k-1)\theta)\right)$, is a **convex function**.

(f) First, notice that ${\rm{dom}}(f)=\mathbb{S}^{n}$ is an affine set (from Lec. 5, p. 1), and therefore, a **convex set**.

From Lec. 9, p. 13, $f(X) = \underset{x^{\top}x=1}{\min}x^{\top}Xx$ over ${\rm{dom}}(f)=\mathbb{S}^{n}$. The affine map $X \mapsto x^{\top}Xx$ is concave in $X\in\mathbb{S}^{n}$ for any fixed vector $x$. Therefore, by Lec. 9, p. 11, the function $f(X)$ being pointwise inf of concave functions, is a **concave function**. 

(g) First, ${\rm{dom}}(f)=\mathbb{R}$ is an affine set, and therefore, a **convex set**.

Next, $f(x) = x - \log(1 + \exp(x))$. Let $g(x):=\log(1 + \exp(x))$. Since $g^{\prime\prime}(x) = \frac{\exp(x)}{(1+\exp(x))^2}>0$, so $g(x)$ is a convex function, and hence $-g(x)$ is a concave function over $\mathbb{R}$. So $f(x) = x + (-g(x))$ is a sum of two concave functions, and therefore, $f(x)$ is a **concave function** over ${\rm{dom}}(f)=\mathbb{R}$. 

Side remark: We can alternatively phrase this result as: ${\rm{logistic}}(x)$ is a log-concave function over the reals.

# Problem 2. [15 + 15 = 30 points] Legendre-Fenchel Conjugates

For each the following function $f$, **derive** the Legendre-Fenchel conjugate $f^{*}$ and its domain ${\rm{dom}}(f^{*})$.

(a) $f(x)=\sum_{i=1}^{n}x_{i}\log x_{i}$ over ${\rm{dom}}(f)=\mathbb{R}^{n}_{++}$.

(b) $f(X) = {\rm{tr}}\left(X^{-1}\right)$ over ${\rm{dom}}(f)=\mathbb{S}^{n}_{++}$.

## solution for Prob. 2

(a) By direct calculation, we find that the function $g(x):=x\log x$ with ${\rm{dom}}(g)=\mathbb{R}_{++}$, has the Legendre-Fenchel conjugate 
$$g^{*}(y)= \left(yx - x\log x\right)\big\vert_{x^{\text{opt}}=\exp(y-1)} = \exp(y-1), \quad {\rm{dom}}(g^{*})=\mathbb{R}.$$ 
Therefore, by Lec. 10, p. 1, the separable sum $f(x)=\sum_{i=1}^{n}g(x_i)$ has the Legendre-Fenchel conjugate
$$f^{*}(y) = \sum_{i=1}^{n}g^{*}(y_i) = \sum_{i=1}^{n}\exp(y_{i}-1), \quad {\rm{dom}}(f^{*})=\mathbb{R}^{n}.$$

(b) By definition, the Legendre-Fenchel conjugate
$$f^{*}(Y) = \underset{X\in{\rm{dom}}(f)=\mathbb{S}^{n}_{++}}{\sup}\bigg\{\langle Y, X\rangle - {\rm{tr}}\left(X^{-1}\right)\bigg\}$$
where $\langle\cdot,\cdot\rangle$ is the Frobenius inner product (see Lec. 10, p. 2-4). Setting the derivative of the above objective w.r.t. $X$ to zero, we find that the maximizer $X^{\text{opt}}$ solves
$$Y + \left(X^{\text{opt}}\right)^{-2} = 0, \quad \Rightarrow\quad X^{\text{opt}} = \left(-Y\right)^{-\frac{1}{2}},$$
where we used that $\frac{\partial}{\partial X}{\rm{tr}}\left(X^{-1}\right) = - X^{-2}$, $X\in\mathbb{S}^{n}_{++}$ (we derived this in Lec. 4, p. 16-17).

Since $X^{\text{opt}}\in\mathbb{S}^{n}_{++}$, the condition $X^{\text{opt}} = \left(-Y\right)^{-\frac{1}{2}}$ requires 
$Y\in\mathbb{S}^{n}_{--}$. 

Finally, substituting $X^{\text{opt}}$ back in the objective, we conclude
$$f^{*}(Y) = \langle Y, X^{\text{opt}}\rangle - {\rm{tr}}\left(\left(X^{\text{opt}}\right)^{-1}\right) = {\rm{tr}}\left(Y\left(-Y\right)^{-\frac{1}{2}}\right) -{\rm{tr}}\left(\left(-Y\right)^{\frac{1}{2}}\right) = -{\rm{tr}}\left((-Y)\left(-Y\right)^{-\frac{1}{2}}\right) -{\rm{tr}}\left(\left(-Y\right)^{\frac{1}{2}}\right) = -2\:{\rm{tr}}\left(\left(-Y\right)^{\frac{1}{2}}\right).$$
Notice that the above expression remains well-defined for $Y\preceq 0$, i.e., when $Y$ is negative semi-definite instead of being strictly negative definite. Therefore, ${\rm{dom}}(f^{*})=\mathbb{S}^{n}_{-}$.

Side remark: The analytical reason (i.e., without appealing to algebra) why $f^{*}(Y)$ remains well defined for negative semi-definite $Y$ is that conjugate functions are always closed, i.e., have closed epigraphs.
